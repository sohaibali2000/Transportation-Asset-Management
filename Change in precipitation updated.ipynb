```python
#Importing all the required libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib 
import seaborn as sns
matplotlib.rcParams["figure.figsize"] = (20,10)

#For ignoring warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the historical climate dataset
df = pd.read_excel('Change in precipitation.xlsx')
```

```python
#display the dataframe
df.head()
```

```python
#get info of our dataframe
df.info()
```

```python
#get summary statistics for data
df.describe()
```

```python
#Check the dataset dimensions
print('Shape of dataset:  ' , df.shape)
```

```python
#check for null values in dataframe
df.isnull().sum()
```

```python
#drop nulls
df.dropna(inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
#lets visualize the distribution of each parameter

f, axes = plt.subplots(2, 2, figsize=(12, 12))
sns.distplot(df['RCP4.5 Precipitation (%) (2005-2050)'] , color="red", ax=axes[0, 0])
sns.distplot(df['RCP4.5 Precipitation (%) (2005-2099)'] , color="olive", ax=axes[0, 1])
sns.distplot(df['RCP8.5 Precipitation (%) (2005-2050)'] , color="green", ax=axes[1, 0])
sns.distplot(df['RCP8.5 Precipitation (%) (2005-2099)'] , color="orange", ax=axes[1, 1])

plt.tight_layout()
```

```python
#plot heatmap
corr = df.corr()

sns.heatmap(corr, cmap='coolwarm', annot=True)
```

```python
# create line chart using seaborn
sns.lineplot(x='County ID', y='RCP4.5 Precipitation (%) (2005-2050)', data=df)

# set the title and axis labels
plt.title('RCP4.5 Precipitation (%) (2005-2050)')
plt.xlabel('County')
plt.ylabel('Change in precipitation')

# show the plot
plt.show()
```
